```python
import numpy as np
import pandas as pd
import re
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

df = kagglehub.load_dataset(KaggleDatasetAdapter.PANDAS,"mohamedbakhet/amazon-books-reviews", path="Books_rating.csv")

print("First 5 records:", df.head())
```

```python
reviews = df['review/text'].tolist()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
```

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return []
    text = text.lower()
    text = re.sub(r"\W+", " ", text)
    words = text.split()
    return [word for word in words if word not in stop_words]

tokenized_reviews = [preprocess_text(review) for review in reviews]
```
